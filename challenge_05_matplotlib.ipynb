## Task 2: Plot a quadratic function

```python
# create a list of x values ranging from -10 to 10
x_values = list(range(-10, 11))
```

```python
# create an empty list for the y values
y_values = []
```

```python
#Â loop over the values of x and calculate y = x^2
for item in x_values:
    y = item ** 2
    y_values.append(y) 
print(y_values)
```

```python
# Plotting the line plot
plt.plot(x_values, y_values, marker='o', linestyle='-')

# Adding labels and title
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('y = x^2')

# Display the plot
plt.show()
```

## Challenge - Visualize the Names Dataset

```python
births_dict = {"year" : [], "total_births" : []}
```

```python
def parse_dataset(value):
    path = f"../data/names/yob{value}.txt"
    df_year = pd.read_csv(path, names=["name", "gender", "frequency"])
    return df_year
```

```python
# Create a plot of the summed up frequency (total births) per year from 1880 to 2020
for year in range(1880, 2021):
    df_year = parse_dataset(year)
    total_births = df_year["frequency"].sum()  # Calculate the sum of the frequencies
    births_dict['total_births'].append(total_births) # Append the result to the dictionary
    births_dict['year'].append(year) # Append the year to the dictionary
yearly_births = pd.DataFrame(births_dict) # Create a yearly_births DataFrame
```

```python
yearly_births
```

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(10,6))
plt.plot(yearly_births["year"],yearly_births["total_births"], color = "steelblue")
plt.title("Total Births per Year")
plt.xlabel("Year")
plt.ylabel("Total Births")
plt.show()
```

## Challenge - Number of unique names over time

```python
# How has the number of unique names given to newborn children varied over time?
# Initialize an empty dictionary to store results
unique_names_dict = {'year': [], 'unique_names_count': []}

# Loop through the years from 1880 to 2020
for year in range(1880, 2021):
    df = parse_dataset(year) # Read in the data for the current year
    unique_names_count = df['name'].nunique() # Calculate the number of unique names for the current year
    unique_names_dict['year'].append(year)    # Append the results to the dictionary
    unique_names_dict['unique_names_count'].append(unique_names_count)

unique_names_df = pd.DataFrame(unique_names_dict) # Convert the dictionary to a Pandas DataFrame

unique_names_df
```

```python
x = unique_names_df["year"]
y = unique_names_df["unique_names_count"]
```

```python
# Visualize the data, use the years on the x-axis and the counts on the y-axis.
plt.figure(figsize=(10,6))
plt.bar(x, y, color=["steelblue"])
plt.title("Number of Unique Names Given to Newborns Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Unique Names")

plt.show()
```
