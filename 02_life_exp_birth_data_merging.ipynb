#Â Merging Population, Birth Rate and Life Expectancy Data Frames

```python
import pandas as pd
```

```python
life_exp = pd.read_csv('../data/life_expectancy_clean.csv')
life_exp.head(1)
```

```python
fert_rate = pd.read_csv('../data/fertility_rate_clean.csv')
fert_rate.head(1)
```

```python
column_to_delete = 'Unnamed: 0'
fert_rate.drop(columns=column_to_delete, inplace=True)
```

```python
fert_rate.head(1)
```

```python
cont = pd.read_csv('../data/continents.csv')
cont.head(1)
```

```python
column_to_delete = 'Unnamed: 0'
cont.drop(columns=column_to_delete, inplace=True)
```

```python
cont.head(1)
```

```python
pop = pd.read_csv('../data/population_clean.csv')
pop.head(1)
```

```python
#Â life expectancy and continents merged
life_cont_merged = life_exp.merge(cont)
life_cont_merged
```

```python
#Â population and fertility merged
pop_fert_merged = pop.merge(fert_rate)
pop_fert_merged
```

```python
df_final = life_cont_merged.merge(pop_fert_merged)
df_final
```

```python
df_final.info()
```

```python
df_final.isna().sum()
```

```python
df_final.to_csv('../data/lifexp_birth_merged.csv')
```
